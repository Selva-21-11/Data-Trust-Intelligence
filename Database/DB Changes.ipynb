```python
%%sql
CREATE TABLE source_transactions_backup AS
SELECT * FROM source_transactions;
```

```python
%%sql
SELECT 
    metric_name,
    record_status,
    COUNT(*) AS row_count
FROM source_transactions
GROUP BY metric_name, record_status
ORDER BY metric_name;
```

```python
%%sql
UPDATE source_transactions
SET record_status = 'missing'
WHERE metric_name = 'Orders'
AND record_status = 'valid'
LIMIT 300;
```

```python
%%sql
UPDATE source_transactions
SET record_status = 'duplicate'
WHERE metric_name = 'Orders'
AND record_status = 'valid'
LIMIT 200;
```

```python
%%sql
UPDATE source_transactions
SET record_status = 'valid'
WHERE metric_name = 'Revenue'
AND record_status = 'missing'
LIMIT 200;
```

```python
%%sql
UPDATE source_transactions
SET record_status = 'valid'
WHERE metric_name = 'Revenue'
AND record_status = 'duplicate'
LIMIT 150;
```

```python
%%sql
UPDATE source_transactions
SET record_status = 'missing'
WHERE metric_name = 'Tickets'
AND record_status = 'valid'
LIMIT 150;
```
