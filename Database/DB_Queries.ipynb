```python
%%sql
SELECT 
    DATE_FORMAT(transaction_date, '%Y-%m') AS month,
    ROUND(SUM(metric_value), 2) AS actual_monthly_revenue
FROM source_transactions
WHERE metric_name = 'Revenue'
  AND record_status = 'valid'
GROUP BY DATE_FORMAT(transaction_date, '%Y-%m')
ORDER BY month;
```

```python
%%sql
SELECT 
    DATE_FORMAT(report_date, '%Y-%m') AS month,
    ROUND(sum(reported_value), 2) AS reported_monthly_revenue
FROM reporting_kpis
WHERE metric_name = 'Revenue'
GROUP BY DATE_FORMAT(report_date, '%Y-%m')
ORDER BY month;
```

```python
%%sql
WITH source_monthly AS (
    SELECT 
        DATE_FORMAT(transaction_date, '%Y-%m') AS month,
        SUM(metric_value) AS actual_revenue
    FROM source_transactions
    WHERE metric_name = 'Revenue'
      AND record_status = 'valid'
    GROUP BY DATE_FORMAT(transaction_date, '%Y-%m')
),
reported_monthly AS (
    SELECT 
        DATE_FORMAT(report_date, '%Y-%m') AS month,
        sum(reported_value) AS reported_revenue
    FROM reporting_kpis
    WHERE metric_name = 'Revenue'
    GROUP BY DATE_FORMAT(report_date, '%Y-%m')
)
SELECT 
    s.month,
    ROUND(s.actual_revenue, 2) AS actual_revenue,
    ROUND(r.reported_revenue, 2) AS reported_revenue,
    ROUND(r.reported_revenue - s.actual_revenue, 2) AS difference,
    ROUND(
        (r.reported_revenue - s.actual_revenue) * 100 / s.actual_revenue,
        2
    ) AS difference_percentage
FROM source_monthly s
JOIN reported_monthly r
ON s.month = r.month
ORDER BY s.month;

```
