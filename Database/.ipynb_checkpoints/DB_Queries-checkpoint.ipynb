```python
%%sql
# CREATE DATABASE data_quality_project;
USE data_quality_project;
```

```python
%%sql
CREATE TABLE source_transactions (
    transaction_id VARCHAR(20),
    customer_id VARCHAR(20),
    source_system VARCHAR(20),
    metric_name VARCHAR(20),
    metric_value DECIMAL(12,2),
    transaction_date DATE,
    record_status VARCHAR(20)
);
```

```python
%%sql
CREATE TABLE reporting_kpis (
    report_id VARCHAR(20),
    metric_name VARCHAR(20),
    reported_value DECIMAL(14,2),
    report_date DATE,
    report_owner VARCHAR(20)
);
```

```python
%%sql
CREATE TABLE data_quality_issues (
    issue_id VARCHAR(20),
    issue_type VARCHAR(20),
    affected_metric VARCHAR(20),
    severity VARCHAR(10),
    detection_date DATE,
    resolution_hours INT
);
```

```python
%%sql
select * from source_transactions
limit 20;
```

```python
%%sql
select record_status, count(record_status) as count
from source_transactions
group by record_status
```

```python
%%sql
SELECT 
    record_status,
    count(*) as record_count,
    count(*) * 100 / SUM(COUNT(*)) OVER () AS percentage
FROM source_transactions
GROUP BY record_status;
```

```python
%%sql
select metric_name,
count(*) as metric_count,
rank() over (order by count(*) desc) as rank_1
from source_transactions
group by metric_name
```

```python
%%sql
select metric_name,
count(*) as duplicate_count
from source_transactions
where record_status = 'duplicate'
group by metric_name
```

```python
%%sql
select metric_name,
count(*) as duplicate_count
from source_transactions
where record_status = 'valid'
group by metric_name
```

```python
%%sql
select * from reporting_kpis
limit 20;
```

```python
%%sql
select metric_name,
round(sum(reported_value),2) as aggregate_value
from reporting_kpis
group by metric_name;
```

```python
%%sql
select metric_name,
round(sum(metric_value),2) as aggregate_value
from source_transactions
group by metric_name;
```
